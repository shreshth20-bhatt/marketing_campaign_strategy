```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
```

```python
ds_train_set['job_type'] = lb.fit_transform(ds_train_set['job_type'])
```

```python
ds_train_set.isnull().any()
```

```python
ds_train_set['income_cat'] = np.ceil(ds_train_set['job_type']/1.5)
```

```python
ds_train_set['income_cat'].where(ds_train_set['income_cat']<5,5.0,inplace=True)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
split= StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)
for train_index, test_index in split.split(ds_train_set,ds_train_set['income_cat']):
    strat_train_set = ds_train_set.loc[train_index]
    strat_test_set = ds_train_set.loc[test_index]
```

```python
for set in (strat_test_set,strat_train_set):
    set.drop('income_cat',axis=1,inplace=True)
```
